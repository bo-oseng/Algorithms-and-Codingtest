### 문제

0보다 크거나 같은 정수 N이 주어진다. 이때, N!을 출력하는 프로그램을 작성하시오.

### 입력

첫째 줄에 정수 N(0 ≤ N ≤ 12)이 주어진다### 출력

### 출력

첫째 줄에 N!을 출력한다.

---

재귀함수를 이용하여 코드를 구현한다.

```python

 # 팩토리얼 함수 선언
def facto(x:int)->int:

    # x 가 1 이하면 1을 출력
    if x<=1:

        return 1

    # x 가 1초과면 x*facto(x-1)을 리턴해 재귀 구현
    else:

        return x*facto(x-1)

nums = int(input())

print(facto(nums))
```
